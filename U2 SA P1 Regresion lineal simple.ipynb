```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv('advertising.csv')
data.head()
```

```python
modelTV = smf.ols('Sales ~ TV', data=data)
modelTV = modelTV.fit()
predTV = modelTV.predict()
plt.figure(figsize=(12, 6))
plt.plot(data['TV'], data['Sales'], 'o')           
plt.plot(data['TV'], predTV, 'r', linewidth=2)
plt.xlabel('Costo de Publicidad')
plt.ylabel('Ventas')
plt.title('TV vs Ventas')

plt.show()
```

```python
new_X = 500
modelTV.predict({"TV": new_X})
```

```python
new_X = 1000
modelTV.predict({"TV": new_X})
```
